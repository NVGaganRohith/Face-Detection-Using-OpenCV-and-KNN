```python
import matplotlib.pyplot as plt
import cv2
```

```python
camera = cv2.VideoCapture(0)
```

```python
success, img = camera.read()

if success:
    img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
    plt.imshow(img)
    
else:
    print("coulnt click the pic")
    
```

```python
camera.release()
```
