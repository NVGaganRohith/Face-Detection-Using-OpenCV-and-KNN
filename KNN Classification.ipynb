```python
#dictionay

mapping = {
    0 : 'Kiwi',
    1 : 'Blackfruit',
}
```

```python
#KNN Algorithm
#creating function
#spaces(indundation) should be given because if not given it will be excluded out of function u are creating 
def dist(p1,p2):
    return np.sum((p1-p2)**2)**.5

def KNN(X,Y,test_point,k=5):
    d = []
    
    #find distance from all points other points
    m = X.shape[0]
    
    for i in range(m):
        current_dist= dist(test_point,X[i])
        d.append((current_dist,Y[i]))
        
    #second step
    d.sort()


    #pick the nearest 5#get clarity
    d = np.array(d)
    labels = d[:5,1]
    uniq,counts = np.unique(labels,return_counts=True)
    print(uniq,counts)

    idx =np.argmax(counts)
    pred = int(uniq[idx])
    return pred
```

```python
pred = KNN(X,Y,test_point)
output = mapping[pred]
print(output)
```
